## <font color=purple>**3. Identify and separate the data into different clusters**</font>

### <font color=blue>*3.a Use Machine Learning algorithms like K Means or Hierarchical clustering to find out the optimal number of clusters in the data.*</font>

### <font color=blue>*K Means*</font>

```python
clusters = range(3,10)
labels_list = []
centeroids_list = []
cluster_errors_list = []

for num in clusters:
    k_means = cluster.KMeans(n_clusters=num,n_init=10,random_state=1)
    k_means.fit(new_dataset_transformed)
    k_means_predictions=k_means.predict(new_dataset_transformed)
    labels = k_means.labels_
    centeroids = k_means.cluster_centers_
    errors = k_means.inertia_
    labels_list.append(labels)
    centeroids_list.append(centeroids)
    cluster_errors_list.append(errors)
```

```python
cluster_df = pd.DataFrame({ "num_clusters":clusters, "cluster_errors": cluster_errors_list} )
```

```python
cluster_df
```

```python
# Elbow plot

plt.figure(figsize=(12,6))
sns.lineplot(x=cluster_df.num_clusters, y=cluster_df.cluster_errors, marker="o")
plt.show()
```

#### <font color=blue>*Here, we observe steep slope till K = 5. After 5, the slope is not much steeper.*</font>
#### <font color=blue>*So, I think we are good with 5 clusters*</font>

```python
k_means = cluster.KMeans(n_clusters=5,n_init=10,random_state=1)
k_means.fit(new_dataset_transformed)
k_means_predictions=k_means.predict(new_dataset_transformed)
```

```python
centeroids = k_means.cluster_centers_
centeroids
```

### <font color=blue>*Few Important Points*</font>
#### <font color=blue>*In K-Means,we choose initial number of clusters, randomly*</font>
#### <font color=blue>*In K-Means,final number of clusters are dependent on initial numbers (of clusters) that were chosen.*</font>
#### <font color=blue>*There can be sub clusters within a cluster which can't be found in K-Means,.*</font>
#### <font color=blue>*In K Means clustering, since we start with random choice of clusters, the results produced by running the algorithm multiple times might differ. While results are reproducible in Hierarchical clustering.*</font>
#### <font color=blue>*High space and time complexity for hierarchical clustering. Hence this clustering algorithm cannot be used when we have huge data.This is because the time complexity of K Means is linear i.e. O(n) while that of hierarchical clustering is quadratic i.e. O(n2).*</font>
#### <font color=blue>*K Means clustering requires prior knowledge of K i.e. no. of clusters you want to divide your data into. But, you can stop at whatever number of clusters you find appropriate in hierarchical clustering by interpreting the dendrogram*</font>
#### <font color=blue>*K Means is found to work well when the shape of the clusters is hyper spherical (like circle in 2D, sphere in 3D). So, hierarchical clustering is used when the variance of distribution of variables is non spherical (one of the assumption of K means is that the distribution of variables should be spherical), it performs relatively better.*</font>

### <font color=blue>*cluster_0_dataset*</font>

```python
cluster_0_dataset_X = cluster_0_dataset.iloc[:,:-1]
cluster_0_dataset_y= cluster_0_dataset.iloc[:,-1:]

cluster_0_dataset_X_train, cluster_0_dataset_X_test, cluster_0_dataset_y_train, cluster_0_dataset_y_test = model_selection.train_test_split(cluster_0_dataset_X, cluster_0_dataset_y, test_size=0.20, random_state=1)

regression_model0 = LinearRegression()

regression_model0.fit(cluster_0_dataset_X_train, cluster_0_dataset_y_train)
cluster_0_dataset_y_predicted = regression_model0.predict(cluster_0_dataset_X_test)


# cluster_0_dataset
train_score  = np.round(regression_model0.score(cluster_0_dataset_X_train,cluster_0_dataset_y_train),2)
test_score   = np.round(regression_model0.score(cluster_0_dataset_X_test,cluster_0_dataset_y_test),2)
mae          = np.round(metrics.mean_absolute_error(cluster_0_dataset_y_test,cluster_0_dataset_y_predicted),7)
print("Train Accuracy ",train_score)
print("Test Accuracy ",test_score)
print("Mean Absolute Error cluster_0_dataset",mae)
mse = np.round(metrics.mean_absolute_error(cluster_0_dataset_y_test,cluster_0_dataset_y_predicted),7)
print("Mean Squared  Error cluster_0_dataset",mse)
r2 = np.round(metrics.r2_score(cluster_0_dataset_y_test,cluster_0_dataset_y_predicted),7)
print("R Square cluster_0_dataset",r2)
print("")
for idx, col_name in enumerate(cluster_0_dataset_X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model0.coef_[0][idx]))
```
