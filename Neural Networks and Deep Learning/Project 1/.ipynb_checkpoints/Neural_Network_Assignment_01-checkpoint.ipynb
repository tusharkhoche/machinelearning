#### <font color=blue>*Splitting Data For Regression*</font>

```python
y = new_dataset.pop('Signal_Strength')
X = new_dataset
```

```python
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.20,random_state=12,stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## <font color=purple>**Part I Regression**</font>

```python
random.seed(7)
```

#### <font color=blue>*Creating ANN Model For Regression*</font>

```python
regression_model = tf.keras.Sequential()
```

##### <font color=blue>Input Layer</font>

```python
regression_model.add(tf.keras.layers.InputLayer(input_shape=(11,)))
```

##### <font color=blue>Normalize the data</font>

```python
regression_model.add(tf.keras.layers.BatchNormalization())
```

##### <font color=blue>Adding Hidden Layers</font>

```python
regression_model.add(tf.keras.layers.Dense(128,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
regression_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
regression_model.add(tf.keras.layers.Dense(64,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
regression_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
regression_model.add(tf.keras.layers.Dense(32,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
regression_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
regression_model.add(tf.keras.layers.Dense(16,kernel_initializer=tf.keras.initializers.HeNormal() ,activation='elu'))
regression_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
regression_model.add(tf.keras.layers.Dense(8,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
regression_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
regression_model.add(tf.keras.layers.Dense(1,activation='linear')) 
```

```python
regression_model.summary()
```

##### <font color=blue>Plot Model does not work on windows. It works only on Google Colab</font>

```python
tf.keras.utils.plot_model(regression_model,show_layer_names=True,show_shapes=True) 
```

##### <font color=blue>Creating Optimizer Object & Compiling the model</font>

```python
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)
```

```python
regression_model.compile(optimizer = adam_optimizer, loss=tf.keras.losses.MSE,metrics='mse') 
```

##### <font color=blue>Creating model checkpoint & callbacks for saving best weights and early stoppage </font>

```python
checkpoint_path = "D:/Python/Jupyter/NeuralNetwork/regression_model-checkpoint/checkpoint.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create a callback that saves the model's weights
callback_best_weights = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1, save_best_only = True, monitor = "val_mse")

callback_early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience = 10,restore_best_weights=True,mode='min')
```

##### <font color=blue>Training the model</font>

```python
regression_model_history = regression_model.fit(X_train, y_train,batch_size=64, epochs=100, verbose=2,validation_data=(X_test, y_test),callbacks=[callback_best_weights,callback_early_stop], use_multiprocessing=True)
```

##### <font color=blue>Evaluating the model</font>

```python
regression_model.evaluate(X_test,y_test,batch_size=64,verbose=1)
```

##### <font color=blue>Printing various metrics using Scikit-Learn</font>

```python
y_predicted = regression_model.predict(X_test)
```

```python
metrics.mean_squared_error(y_true=y_test,y_pred=y_predicted)
```

```python
metrics.r2_score(y_true=y_test,y_pred=y_predicted)
```

```python
metrics.mean_absolute_error(y_true=y_test,y_pred=y_predicted)
```

##### <font color=blue>Saving the model</font>

```python
regression_model.save(filepath="D:/Python/Jupyter/NeuralNetwork/regression_model", overwrite=True, include_optimizer=True)
```

##### <font color=blue>Plotting train & test graph vs mse</font>

```python
plt.plot(regression_model_history.history['mse'])
plt.plot(regression_model_history.history['val_mse'])
plt.title('model mse')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(regression_model_history.history['loss'])
plt.plot(regression_model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

## <font color=purple>**Part II Classification**</font>

#### <font color=blue>*Preparing data*</font>

```python
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.20,random_state=12,stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=9)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=9)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
print(y_train[0])
```

##### <font color=blue>*Creating ANN Model For Classification*</font>

```python
classification_model = tf.keras.Sequential()
```

##### <font color=blue>Input Layer</font>

```python
classification_model.add(tf.keras.layers.InputLayer(input_shape=(11,)))
```

##### <font color=blue>Normalize the data</font>

```python
classification_model.add(tf.keras.layers.BatchNormalization())
```

##### <font color=blue>Adding Hidden Layers</font>

```python
classification_model.add(tf.keras.layers.Dense(288,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
classification_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
classification_model.add(tf.keras.layers.Dense(144,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
#classification_model.add(tf.keras.layers.ELU())
classification_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
classification_model.add(tf.keras.layers.Dense(72,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
#classification_model.add(tf.keras.layers.ELU())
classification_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
classification_model.add(tf.keras.layers.Dense(36,kernel_initializer=tf.keras.initializers.HeNormal() ,activation='elu'))
#classification_model.add(tf.keras.layers.ELU())
classification_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
classification_model.add(tf.keras.layers.Dense(18,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
#classification_model.add(tf.keras.layers.ELU())
classification_model.add(tf.keras.layers.Dropout(rate=0.10))
```

```python
classification_model.add(tf.keras.layers.Dense(9,activation='softmax')) 
```

```python
classification_model.summary()
```

```python
classification_model.compile(optimizer = adam_optimizer, loss='categorical_crossentropy',metrics=['accuracy'])  
```

##### <font color=blue>Creating model checkpoint & callbacks for saving best weights and early stoppage </font>

```python
checkpoint_path = "D:/Python/Jupyter/NeuralNetwork/classification_model-checkpoint/checkpoint.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
# Create a callback that saves the classification_model's weights
callback_best_weights = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1, save_best_only = True, monitor = "val_accuracy")

callback_early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience = 10,restore_best_weights=True,mode='min')
```

```python
classification_model_history = classification_model.fit(X_train, y_train,batch_size=64, epochs=100, verbose=1,validation_data=(X_test, y_test),callbacks=[callback_best_weights,callback_early_stop], use_multiprocessing=True) 
```

```python
def doneImportData():
    
    confirm = "Done"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=0,column=8)
    ConfirmEntrybox.insert(1,str(confirm)) 
    
def noExtension():
    
    confirm = "No Extension"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=0,column=8)
    ConfirmEntrybox.insert(1,str(confirm))     

def importData():
    try:
        global dataset
        dataset_name = Step01DatasetName.get()
        dataset_extension = re.findall("\..*",dataset_name)
        print("dataset_name-"+dataset_name)
                
        if(dataset_extension==[".csv"]):
            missing_value_formats = ["n.a.","?","NA","n/a","na","N.A.","nan","NAN","--"," ", "  ","@",".","-"]
            dataset = pd.read_csv(dataset_name, na_values = missing_value_formats)  
            doneImportData()
        elif(dataset_extension==[".xlsx"]):
            dataset = pd.read_excel(dataset_name)
            doneImportData()
        elif(dataset_extension==[".xls"]):
            dataset = pd.read_excel(dataset_name)    
            doneImportData()
        elif(dataset_extension==[""]):    
            noExtension()

    except FileNotFoundError:
        confirm = "No Such File"
        ConfirmEntrybox=ttk.Entry(win,width=25)
        ConfirmEntrybox.grid(row=0,column=8)
        ConfirmEntrybox.insert(1,str(confirm))
        

def targetFound():
    
    found = "Found"
    foundEntrybox=ttk.Entry(win,width=25)
    foundEntrybox.grid(row=2,column=8)
    foundEntrybox.insert(1,str(found)) 
    
    
def importTarget():
    try:
        global targetColumnName
        targetColumnName = Step02TargetColumnName.get()
        if(targetColumnName in dataset.columns):
            targetFound()
       
    except KeyError: 
        noTarget = "Incorrect Target "+targetColumnName
        noTargetEntrybox=ttk.Entry(win,width=40)
        noTargetEntrybox.grid(row=2,column=8)
        noTargetEntrybox.insert(1,str(noTarget))
```

```python
def cleanAndPreprocessDataForRegression():
    importData()
    dataset.drop_duplicates(inplace=True)
    y = dataset.pop(targetColumnName)
    X = dataset
    
    
    global X_train
    global X_test
    global y_train
    global y_test
    
    X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.20,random_state=7,stratify=y)
    print(X_train.shape)
    print(X_test.shape)
    print(y_train.shape)
    print(y_test.shape)    
```

```python
def regressorModelEvaluate():
    regression_model.evaluate(X_test,y_test,batch_size=64,verbose=2)
```

```python
def trainRegressor():
    cleanAndPreprocessDataForRegression()
    
    global regression_model
    
    regression_model = tf.keras.Sequential()
    regression_model.add(tf.keras.layers.InputLayer(input_shape=(11,)))
    regression_model.add(tf.keras.layers.BatchNormalization())
    regression_model.add(tf.keras.layers.Dense(128,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    regression_model.add(tf.keras.layers.Dropout(rate=0.10))
    regression_model.add(tf.keras.layers.Dense(64,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    regression_model.add(tf.keras.layers.Dropout(rate=0.10))
    regression_model.add(tf.keras.layers.Dense(32,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    regression_model.add(tf.keras.layers.Dropout(rate=0.10))
    regression_model.add(tf.keras.layers.Dense(16,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    regression_model.add(tf.keras.layers.Dropout(rate=0.10))
    regression_model.add(tf.keras.layers.Dense(8,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    regression_model.add(tf.keras.layers.Dropout(rate=0.10))
    regression_model.add(tf.keras.layers.Dense(1,activation='linear')) 
    regression_model.summary()

    #Plot Model does not work on windows. It works only on Google Colab
    #tf.keras.utils.plot_model(regression_model,show_layer_names=True,show_shapes=True) 
    
    adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    regression_model.compile(optimizer = adam_optimizer, loss=tf.keras.losses.MSE,metrics='mse') 

    checkpoint_path = "D:/Python/Jupyter/NeuralNetwork/regression_model-checkpoint/checkpoint.ckpt"
    checkpoint_dir = os.path.dirname(checkpoint_path)

    # Create a callback that saves the model's weights
    callback_best_weights = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                     save_weights_only=True,
                                                     verbose=1, save_best_only = True, monitor = "val_mse")

    callback_early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience = 10,restore_best_weights=True,mode='min')

    regression_model_history = regression_model.fit(X_train, y_train,batch_size=64, epochs=70, verbose=1,validation_data=(X_test, y_test),callbacks=[callback_best_weights,callback_early_stop], use_multiprocessing=True)

        
    
    y_predicted = regression_model.predict(X_test)
    print("mean_squared_error->",metrics.mean_squared_error(y_true=y_test,y_pred=y_predicted))
    print("r2_score->",metrics.r2_score(y_true=y_test,y_pred=y_predicted))
    print("mean_absolute_error",metrics.mean_absolute_error(y_true=y_test,y_pred=y_predicted))

    plt.plot(regression_model_history.history['mse'])
    plt.plot(regression_model_history.history['val_mse'])
    plt.title('model mse')
    plt.ylabel('mse')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(regression_model_history.history['loss'])
    plt.plot(regression_model_history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

    regressorModelEvaluate()
    donetrainRegressor()

def donetrainRegressor():
    
    confirm = "Regressor Model Trained"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=5,column=3)
    ConfirmEntrybox.insert(1,str(confirm)) 
    
```

```python
def saveRegressor():
    regression_model.save(filepath="D:/Python/Jupyter/NeuralNetwork/regression_model", overwrite=True, include_optimizer=True)
    doneSaveRegressor()

def doneSaveRegressor():
    
    confirm = "Regressor Model Saved"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=7,column=3)
    ConfirmEntrybox.insert(1,str(confirm))

```

```python
def cleanAndPreprocessDataForClassification():
    importData()
    dataset.drop_duplicates(inplace=True)
    y = dataset.pop(targetColumnName)
    X = dataset
    
    print("cleanAndPreprocessDataForClassification",X.shape)
    print("cleanAndPreprocessDataForClassification",y.shape)
    
    global X_train
    global X_test
    global y_train
    global y_test
    
    X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.20,random_state=7,stratify=y)
    
    print(X_train.shape)
    print(X_test.shape)
    print(y_train.shape)
    print(y_test.shape)    
    
    y_train = tf.keras.utils.to_categorical(y_train, num_classes=9)
    y_test = tf.keras.utils.to_categorical(y_test, num_classes=9)
```

```python
def classificationModelEvaluate():
    classification_model.evaluate(X_test,y_test,batch_size=64,verbose=2)
```

```python
def trainClassifier():
    cleanAndPreprocessDataForClassification()
    
    global classification_model
    
    classification_model = tf.keras.Sequential()
    #Input Layer
    classification_model.add(tf.keras.layers.InputLayer(input_shape=(11,)))
    # Normalize the data
    classification_model.add(tf.keras.layers.BatchNormalization())
    classification_model.add(tf.keras.layers.Dense(288,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    classification_model.add(tf.keras.layers.Dropout(rate=0.15))
    
    classification_model.add(tf.keras.layers.Dense(144,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    classification_model.add(tf.keras.layers.Dropout(rate=0.15))
    
    classification_model.add(tf.keras.layers.Dense(72,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    classification_model.add(tf.keras.layers.Dropout(rate=0.15))
    
    classification_model.add(tf.keras.layers.Dense(36,kernel_initializer=tf.keras.initializers.HeNormal() ,activation='elu'))
    classification_model.add(tf.keras.layers.Dropout(rate=0.15))
    
    classification_model.add(tf.keras.layers.Dense(18,kernel_initializer=tf.keras.initializers.HeNormal(),activation='elu'))
    classification_model.add(tf.keras.layers.Dropout(rate=0.15))
    
    classification_model.add(tf.keras.layers.Dense(9,activation='softmax')) 
    
    classification_model.summary()
    
    #Plot Model does not work on windows. It works only on Google Colab
    #tf.keras.utils.plot_model(classification_model,show_layer_names=True,show_shapes=True) 
    
    adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
    
    classification_model.compile(optimizer = adam_optimizer, loss='categorical_crossentropy',metrics=['accuracy'])  

    checkpoint_path = "D:/Python/Jupyter/NeuralNetwork/classification_model-checkpoint/checkpoint.ckpt"
    checkpoint_dir = os.path.dirname(checkpoint_path)

    # Create a callback that saves the classification_model's weights
    callback_best_weights = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                     save_weights_only=True,
                                                     verbose=1, save_best_only = True, monitor = "val_accuracy")

    callback_early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience = 10,restore_best_weights=True,mode='min')


    classification_model_history = classification_model.fit(X_train, y_train,batch_size=64, epochs=50, verbose=1,validation_data=(X_test, y_test),callbacks=[callback_best_weights,callback_early_stop], use_multiprocessing=True) 

    

    plt.plot(classification_model_history.history['accuracy'])
    plt.plot(classification_model_history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(classification_model_history.history['loss'])
    plt.plot(classification_model_history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

    classificationModelEvaluate()
    donetrainClassifier()

    

    
def donetrainClassifier():
    
    confirm = "Classifier Model Trained"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=9,column=3)
    ConfirmEntrybox.insert(1,str(confirm)) 
```

```python
def saveClassifier():
    classification_model.save(filepath="D:/Python/Jupyter/NeuralNetwork/classification_model", overwrite=True, include_optimizer=True)
    doneSaveClassifier()

def doneSaveClassifier():
    
    confirm = "Classifier Model Saved"
    ConfirmEntrybox=ttk.Entry(win,width=25)
    ConfirmEntrybox.grid(row=11,column=3)
    ConfirmEntrybox.insert(1,str(confirm))
```
